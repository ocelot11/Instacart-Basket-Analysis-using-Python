#03. Using transform() and loc() functions

```python
# Using the transform() function to create the "max_order" column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(20)
```

```python
# Using the loc() functions to create the customer loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

#04. Spending habits of the three types of customers

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

#05. Finding the different types of spenders

```python
# Using the transform() function to create the "mean_price_of_products" column
ords_prods_merge['mean_price_of_products'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking the aggregation results
ords_prods_merge.head(20)
```

```python
# Checking the aggregation results for one user
ords_prods_merge[ords_prods_merge['user_id'] == 19]
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price_of_products'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price_of_products'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

#05. Finding the regularity of a user's ordering behaviour

```python
# Using the transform() function to create the "mean_days_since_prior_order" column
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge['median_days_since_prior_order'].value_counts(dropna = False)
```

```python
# Using the loc() function to create the "Regularity_of_customer" column
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'Regularity_of_customer'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'Regularity_of_customer'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'Regularity_of_customer'] = 'Frequent customer'
```

```python
ords_prods_merge['Regularity_of_customer'].value_counts(dropna = False)
```
