### 02. Security implications - Looking for Personally identifiable information (PII) data

```python
ords_prods_all.info()
```

The ords_prods_custs_merge data set has some sensitive information in our data set that could lead to the identification of an individual - First name, surname, state, age, number_of_dependents, marital_status and income. Therfore, we are working with PII. 
In order to comply with European Union's data privacy regulation, GDPR, the columns First name and surname will be dropped.

```python
# Dropping columns 'First name' and 'surname'
col_drop = ['First Name', 'surname']
ords_prods_all = ords_prods_all.drop(columns = col_drop)
```

```python
def new_prices(row):
    if row['prices'] >= 1 and row['prices'] <= 2.5:
        return 2.5    
    elif row['prices'] >= 2.6 and row['prices'] <= 5:
        return 5
    elif row['prices'] >= 5.1 and row['prices'] <= 7.5:
        return 7.5
    elif row['prices'] >= 7.6 and row['prices'] <= 10:
        return 10
    elif row['prices'] >= 10.1 and row['prices'] <= 12.5:
        return 12.5
    elif row['prices'] >= 12.6 and row['prices'] <= 15:
        return 15
    elif row['prices'] >= 15.1 and row['prices'] <= 17.5:
        return 17.5
    elif row['prices'] >= 17.6 and row['prices'] <= 20:
        return 20
    elif row['prices'] >= 20.1 and row['prices'] <= 22.5:
        return 22.5
    elif row['prices'] >= 22.6 and row['prices'] <= 25:
        return 25
```

```python
ords_prods_high2['new_prices'] = ords_prods_high2.apply(new_prices, axis=1)
```

```python
def get_status_flag(row):
    if row['age'] > 21 and row['marital_status'] == 'married' and row['income'] > 1000:
        return 'PARENT'
    else:
        return None

ords_prods_all['STATUSFLAG'] = ords_prods_all.apply(get_status_flag, axis=1)
```
